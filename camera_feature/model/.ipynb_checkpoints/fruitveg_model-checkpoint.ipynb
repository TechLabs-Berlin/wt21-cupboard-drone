# Fruit and vegtable classifier

In this notebook I train an image classifier with the objective of recognising a variety of vegetables and fruits for our web app **The Cupboard** from an image uploaded by the user.

I used the **Grocery Store image dataset** created by Marcus Klasson, which uses natural images of grocery items. I removed the pakckaged items folder to concentrate on fruit and vegtable classification. Photographs of packages were taken in Sweden, which would affect the accuracy of the model when applied to images of German products and in our case it makes sense to implement a barcode reader to identify packaged products.

The fruit and vegtables dataset and associated paper can be found at:  

https://paperswithcode.com/paper/a-hierarchical-grocery-store-image-dataset  

https://github.com/marcusklasson/GroceryStoreDataset

```python
from fastbook import *
import fastai
from fastai import *
fastai.__version__
```

```python
from fastai.vision.all import *
```

```python
path = Path('gdrive/MyDrive/Colab Notebooks/Cupboard/GroceryStoreDataset/dataset/train')
```

```python
Path.BASE_PATH = path
path.ls()
```

```python
fns = get_image_files(path)
fns
```

```python
failed = verify_images(fns)
failed
```

```python
fruitveg = DataBlock(blocks = (ImageBlock, CategoryBlock),
                 get_items=get_image_files, 
                 splitter=RandomSplitter(valid_pct=0.2, seed=42),
                 get_y=parent_label,
                 item_tfms=Resize(460),
                 batch_tfms=aug_transforms(size=224, min_scale=0.5))
```

```python
dls = fruitveg.dataloaders(path)
```

```python
dls.valid.show_batch(max_n=9, nrows=3)
```

```python
dls.train.show_batch(max_n=4, nrows=1, unique=True)
```

Training a computer vision model requires a lot of labeled data to achieve a decent result.
For this reason, I opted for using a pretrained CNN such as a ResNet. An architecture with 34 layers
was quite successful in recognizing the fruits and vegetables in our dataset.

```python
learn = cnn_learner(dls, resnet34, metrics=error_rate)
learn.fine_tune(4)
```

```python
interp = ClassificationInterpretation.from_learner(learn)
interp.plot_confusion_matrix(figsize=(12,12), dpi=60)
```

```python
interp.most_confused(min_val=5)
```

```python
learn.export()
```

```python
path = Path()
path.ls(file_exts='.pkl')
```
